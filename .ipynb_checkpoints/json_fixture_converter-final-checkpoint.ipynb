```python
import json
from datetime import datetime
```

```python
with open('data.json') as json_data:
    content = json.load(json_data)
```

```python
content
```

```python
list_length = len(content)
```

```python
list_length
```

```python
# CREATE THIRD ELEMENT FIELDS
# "fields" ; { ... }
# Includes date formatting YYYY-MM-DD
```

```python
fields_list = []
for x in range(0, list_length - 1):
    fields_dict = {
        "first_name" : content[x]['first_name'],
        "last_name" : content[x]["last_name"],
        "email" : content[x]["email"],
        "gender" : content[x]["gender"],
        "date_of_birth" : datetime.strptime(content[x]["date_of_birth"], '%d/%M/%Y').strftime('%Y-%m-%d'),
        "industry" : content[x]["industry"],
        "salary" : content[x]["salary"],
        "years_of_experience" : content[x]["years_of_experience"]
    }
    fields_list.append(fields_dict)
```

```python
# ADD ALL ELEMENTS TO FIXTURE DATA
# "model" : "api.employee",
# "pk" : x,  1-2999
# "fields" : fields_list[x]
```

```python
fixture_data = []
```

```python
for x in range(0, list_length - 1):
    fixture_data.append({"model":"api.employee", "pk": int(x + 1), "fields" : fields_list[x]})
    
```

```python
fixture_data
```

```python
# SAVE ON JSON FILE
```

```python
jsonString = json.dumps(fixture_data)
jsonFile = open("fixture_data.json", "w")
jsonFile.write(jsonString)
jsonFile.close()
```
